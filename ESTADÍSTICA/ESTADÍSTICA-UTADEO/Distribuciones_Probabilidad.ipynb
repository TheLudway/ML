# Cosas importantes de las distribuciones de probabilidad

## Librerías

```python
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy.stats import binom
```

## Distribuciones gráficamente


```python
n = 20
p = 0.25
x = np.arange(0, n+1)
pmf = np.zeros(n+1)
for i in range(n + 1):
    pmf[i] = math.comb(n, i) * p**i * (1-p)**(n-i)
fig, ax = plt.subplots(figsize=(5, 2))
plt.stem(x, pmf)
ax.set_xlim(0, n)
ax.xaxis.set_major_locator(plt.MultipleLocator(1))
plt.title(f'Distribución binomial con, $n={n}$ y $p={p}$')
plt.xlabel('Número de acontecimientos')
plt.ylabel('Probabilidad')
plt.show()
```
